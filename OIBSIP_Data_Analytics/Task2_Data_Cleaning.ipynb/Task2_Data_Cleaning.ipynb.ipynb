# Data Cleaning on Titanic Dataset


## Objective

The objective of this project is to clean the Titanic dataset by identifying and handling missing values, removing duplicate records, checking data types, detecting outliers, and preparing the dataset for further analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
```

### Observation

The dataset contains missing values in columns such as Age, Cabin, and Embarked. These missing values need to be handled before performing further analysis.

```python
df.duplicated().sum()
```

### Observation

The dataset was checked for duplicate records. If duplicates are found, they will be removed to improve data quality.

```python
df.drop_duplicates(inplace=True)

print("Duplicates removed successfully.")
```

```python
df.isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())

df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

df["Cabin"] = df["Cabin"].fillna("Unknown")
```

```python
df.isnull().sum()
```

### Observation

Initially, the dataset contained missing values in the Age, Cabin, and Embarked columns. These were handled by replacing missing Age values with the median, Embarked values with the mode, and Cabin values with "Unknown". After cleaning, the dataset contained no missing values.

```python
df.duplicated().sum()
```

### Observation

The dataset was checked for duplicate records, and no duplicate rows were found.

```python
df.drop_duplicates(inplace=True)
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df["Age"])
plt.title("Box Plot of Age")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df["Fare"])
plt.title("Box Plot of Fare")
plt.show()
```

```python
df.to_csv("Cleaned_Titanic.csv", index=False)

print("✅ Cleaned dataset saved successfully!")
```
